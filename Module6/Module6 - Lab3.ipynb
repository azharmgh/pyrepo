```python
from sklearn import preprocessing

def preprocessandScore(processor,xtrain,ytrain,xtest,ytest, Cs, gs):
    pm = processor.fit(xtrain)
    Xtrain = pm.transform(xtrain)
    Xtest = pm.transform(xtest)
    bestscore = 0.0
    best_c = 0
    best_gamma = 0

    for c in (Cs):
        for g in (gs):
            model = SVC(C=c,gamma=g)
            model.fit(Xtrain,y_train['status'])
            score = model.score(Xtest,y_test['status'])
            if score > bestscore:
                bestscore = score
                best_c = c
                best_gamma = g

    print('processor is ', type(processor))
    print("bestscore is = ", bestscore)
    print("best c is = ", best_c)
    print("best gamma is = ", best_gamma)


    return model
```

```python
from sklearn.decomposition import PCA

def preprocessAndScoreWithPCA(processor,xtrain,ytrain,xtest,ytest, Cs, gs, pcas):
    pm = processor.fit(xtrain)
    Xtrain = pm.transform(xtrain)
    Xtest = pm.transform(xtest)
    bestscore = 0.0
    best_c = 0
    best_gamma = 0
    best_components = 0
    
    for p in (pcas):
        pca = PCA(n_components=p, svd_solver='randomized', random_state=7)
        pca.fit(Xtrain)
        XTtrain = pca.transform(Xtrain)
        XTtest = pca.transform(Xtest)
        for c in (Cs):
            for g in (gs):
                model = SVC(C=c,gamma=g)
                model.fit(XTtrain,y_train['status'])
                score = model.score(XTtest,y_test['status'])
                if score > bestscore:
                    bestscore = score
                    best_c = c
                    best_gamma = g
                    best_components = p

    print('processor is ', type(processor))
    print("bestscore is = ", bestscore)
    print("best c is = ", best_c)
    print("best gamma is = ", best_gamma)
    print("best pca components are: ", best_components)


    return model
```

```python
preprocessAndScoreWithPCA(preprocessing.StandardScaler(),X_train,y_train,X_test,y_test,crange,gammarange,pcarange)
```

```python
nneighbours = np.arange(2,6,1)
ncomponents = np.arange(4,7,1)
```

```python
from sklearn import manifold


def preprocessAndScoreWithISOMAP(processor,xtrain,ytrain,xtest,ytest, Cs, gs, neighbours,components):
    pm = processor.fit(xtrain)
    Xtrain = pm.transform(xtrain)
    Xtest = pm.transform(xtest)
    bestscore = 0.0
    best_c = 0
    best_gamma = 0
   
    best_n_neighbours = 0
    best_n_components = 0
    
    for n in (neighbours):
        for cmp in (components):
            iso = manifold.Isomap(n, cmp)
            iso.fit(Xtrain)
            XTtrain = iso.transform(Xtrain)
            XTtest = iso.transform(Xtest)
            for c in (Cs):
                for g in (gs):
                    model = SVC(C=c,gamma=g)
                    model.fit(XTtrain,y_train['status'])
                    score = model.score(XTtest,y_test['status'])
                    if score > bestscore:
                        bestscore = score
                        best_c = c
                        best_gamma = g
                        best_n_neighbours = n
                        best_n_components = cmp

    print('processor is ', type(processor))
    print("bestscore is = ", bestscore)
    print("best c is = ", best_c)
    print("best gamma is = ", best_gamma)
    print("best iso components are: ", best_n_components)
    print("best iso neighbours are:",  best_n_neighbours)


    return model
```

```python
preprocessAndScoreWithISOMAP(preprocessing.StandardScaler(),X_train,y_train,X_test,y_test,crange,gammarange,nneighbours,ncomponents)
```
